###**INCREMENTAL LOADING USING PARAMETERS AND VARIABLES**

```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

```python
dbutils.widgets.text("filename"," ")
```

```python
filename = dbutils.widgets.get("filename")
filename
```

```python
df = spark.readStream.format("cloudFiles")\
                      .option("cloudFiles.format", "csv")\
                      .option("CloudFiles.schemaLocation", f"/Volumes/streaming/bronze/bronzevolume/{filename}/checkpoint")\
                      .option("cloudFiles.schemaEvolutionMode", "rescue")\
                      .load(f"/Volumes/streaming/raw/rawvolume/{filename}/")
```

```python
# ingesting raw data from source to bronze layer using lakeflow jobs
df.writeStream.format("delta")\
              .outputMode("append")\
              .trigger(once=True)\
              .option("checkpointLocation", f"/Volumes/streaming/bronze/bronzevolume/{filename}/checkpoint")\
              .option("path",f"/Volumes/streaming/bronze/bronzevolume/{filename}/data")\
              .start()
```
